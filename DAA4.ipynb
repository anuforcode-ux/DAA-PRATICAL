```python
import heapq


def dijkstra(graph, source):
    """
    Dijkstra's Algorithm using Min-Heap

    Time: O((V + E) log V)
    Space: O(V)

    graph: dict {u: [(v, weight), ...]}
    """

    n = len(graph)

    dist = [float('inf')] * n
    prev = [None] * n

    dist[source] = 0

    # (distance, vertex)
    pq = [(0, source)]

    visited = set()

    while pq:

        d, u = heapq.heappop(pq)

        # Skip if vertex was already processed
        if u in visited:
            continue

        visited.add(u)

        # Check all neighbouring vertices
        for v, w in graph[u]:

            # Relaxation
            if dist[u] + w < dist[v]:

                dist[v] = dist[u] + w

                prev[v] = u

                heapq.heappush(
                    pq,
                    (dist[v], v)
                )

    return dist, prev


def reconstruct_path(prev, source, target):

    path = []

    node = target

    while node is not None:

        path.append(node)

        node = prev[node]

    path.reverse()

    # Check whether source is actually
    # the starting point of the path
    if path[0] == source:
        return path

    return []


# --------------------------------
# Graph Definition
# --------------------------------

graph = {
    0: [(1, 6), (2, 3)],
    1: [(3, 2), (4, 5)],
    2: [(1, 2), (3, 7), (4, 4)],
    3: [(4, 1), (5, 3)],
    4: [(5, 2), (6, 6)],
    5: [(6, 1)],
    6: []
}


# --------------------------------
# Source Vertex
# --------------------------------

source = 0


# --------------------------------
# Run Dijkstra
# --------------------------------

dist, prev = dijkstra(graph, source)


# --------------------------------
# Print Results
# --------------------------------

print(f"Shortest paths from vertex {source}:")

print(
    f'{"Vertex":>8} '
    f'{"Distance":>10} '
    f'{"Path":>30}'
)

print('-' * 55)


for v in range(len(graph)):

    path = reconstruct_path(
        prev,
        source,
        v
    )

    if path:
        path_str = ' -> '.join(
            map(str, path)
        )
    else:
        path_str = 'No path'

    print(
        f'{v:>8} '
        f'{dist[v]:>10} '
        f'{path_str:>30}'
    )
```
